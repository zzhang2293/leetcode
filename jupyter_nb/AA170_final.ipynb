```python
import numpy as np
import pandas as pd
def column_label_to_number(column_label):
    number = 0
    for i, char in enumerate(reversed(column_label)):
        number += (ord(char.upper()) - ord('A') + 1) * (26 ** i)
    return number - 1  # Subtract 1 to make it zero-indexed

"""
    A B C ... EM ... FA _ FB ... FO _ _... 
    FB - FO -> EM - FA - 1
    FA -> _ 

"""
df = pd.read_csv("files.csv")
df_cp = pd.DataFrame()
start = column_label_to_number("EM")
end = column_label_to_number("FA")
repli_start = column_label_to_number("FB") + 1 # plus one because we insert a new column
repli_end = column_label_to_number("FO") + 1
df.insert(loc=end + 1, column="area", value=np.nan)
for index, row_data in df.iterrows():
    modified = row_data.drop(df.columns[repli_start:repli_end + 1])
    df_cp = pd.concat([df_cp, pd.DataFrame(modified).T])
    if not pd.isna(row_data.iloc[repli_start]):
        row_data.iloc[start:end] = row_data.iloc[repli_start:repli_end + 1]
        row_data.iloc[end] = np.nan
        row_data.iloc[end + 1] = 528
        modified = row_data.drop(df.columns[repli_start:repli_end + 1])
        df_cp = pd.concat([df_cp, pd.DataFrame(modified).T])
df_cp.to_csv("files_cp.csv", index=False)
```
